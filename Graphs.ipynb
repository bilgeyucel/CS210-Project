```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Movies.csv',index_col=['releaseYear'])
d = data.dropna()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(range(1,13),['January','February','March','April','May','June',
            'July','August','September','October','November','December'])
data.groupby(['releaseMonth']).size().plot(c='r')
plt.ylabel('Count')
plt.xlabel('Months')
```

```python
plt.figure(figsize=(15,7))
plt.xticks(range(1,14),['Action','Adevnture','Animation','Comedy','Crime','Drama',
            'Fantasy','History','Horror','Religious','Romance','Science - Fiction', 'Thriller'])
data.groupby(['Genre_Enum']).size().plot(c='y')
plt.ylabel('Count')
plt.xlabel('Genres')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(17,8))
ax = fig.add_subplot(111)

plt.scatter(data.Genre_Enum, data.ticketsSold)
plt.xticks(range(1,14),['Action','Adevnture','Animation','Comedy','Crime','Drama',
            'Fantasy','History','Horror','Religious','Romance','Science - Fiction', 'Thriller'])
plt.xlabel('Genre')
plt.ylabel('Sold Tickets')
```
